## Plots

```python
sns.countplot(data=df, x="survived")
plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(data=df, x="sex", hue="survived")
plt.title("Survival by Sex")
plt.show()
```

```python
df["age"].hist(bins=30)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```
